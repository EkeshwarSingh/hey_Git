```python
np.zeros(5)        # [0. 0. 0. 0. 0.]
np.ones((2, 3))    # 2x3 matrix of ones
np.arange(0, 10)   # 0 to 9
np.linspace(0, 1, 5)  # 5 equally spaced values
```

```python
arr2 = np.arange(1, 100, 5)
print(arr2)
print(arr2.shape)
```
