```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'City' : ['Paris', 'London', 'Dubai', 'Mumbai'],
    'Return flight' : [200, 250, 370, 450],
    'Hotel per day' : [20, 30, 15, 10],
    'Weekly car rental' : [200, 120, 80, 70]
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
def Total_cost(Return_flight, Hotel_per_day, Weekly_car_rental):
    Total_cost = Return_flight + Hotel_per_day*7 + Weekly_car_rental
    return
```

```python
df['Total cost'] = (
    df['Return flight'] +
    (df['Hotel per day'] * 7) +
    df['Weekly car rental']
)

print(df)
```

```python
min_cost = df['Total cost'][0]
cheapest_city = df['City'][0]

for i in range(len(df)):
    if df['Total cost'][i] < min_cost:
        min_cost = df['Total cost'][i]
        cheapest_city = df['City'][i]

print("Cheapest city:", cheapest_city)
print("Minimum cost for 1 week:", min_cost)
```

```python
min_index = df['Total cost'].idxmin()

print("Cheapest city:", df.loc[min_index, 'City'])
print("Minimum cost for 1 week:", df.loc[min_index, 'Total cost'])
```

```python
import pandas as pd

data = {
    'City' : ['Paris', 'London', 'Dubai', 'Mumbai'],
    'Return flight' : [200, 250, 370, 450],
    'Hotel per day' : [20, 30, 15, 10],
    'Weekly car rental' : [200, 120, 80, 70]
}
df = pd.DataFrame(data)

# Replicate the effect of cost_calculator with Days=10, as seen in previous cell execution
Days = int(input("How many days?"))
df['cost'] = df['Return flight'] + (df['Hotel per day'] * Days) + df['Weekly car rental']

min_cost = df['cost'][0]
cheapest_city = df['City'][0]

for i in range(len(df)):
    if df['cost'][i] < min_cost:
        min_cost = df['cost'][i]
        cheapest_city = df['City'][i]

print("Cheapest city:", cheapest_city)
print("Minimum cost for days:", min_cost)
```
